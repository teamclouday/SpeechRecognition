```python
import random
def random_check(df):
    rnd_idx = random.randint(0, len(df)-1)
    with open(df["Path"][rnd_idx], "rb") as inFile:
        wav, sr = pickle.load(inFile)
    print(df["Text"][rnd_idx])
    return ipd.Audio(wav, rate=sr)
```

```python
random_check(data_train)
```

```python
random_check(data_test)
```
